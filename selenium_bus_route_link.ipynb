```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException, TimeoutException, ElementNotInteractableException
import time
import pandas as pd

# Initialize WebDriver
driver_A = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
driver_A.maximize_window()

# Open the desired webpage
driver_A.get("https://www.redbus.in/online-booking/ksrtc-kerala/?utm_source=rtchometile")
time.sleep(3)  # Allow time for the page to load


# Function to retrieve bus links and routes
def kerala_link_route(path):
    LINKS_KERALA = []
    ROUTE_KERALA = []
    
    # WebDriverWait instance
    wait_A = WebDriverWait(driver_A, 10)
    
    # Retrieve route links from multiple pages
    while True:
        try:
            # Extract route links and texts
            paths = driver_A.find_elements(By.XPATH, path)
            for links in paths:
                d = links.get_attribute("href")
                if d:
                    LINKS_KERALA.append(d)
            for route in paths:  # Corrected indentation
                ROUTE_KERALA.append(route.text)  # Added the correct line inside the loop
            try:
                # Get the active page number
                active_page_element = driver_A.find_element(By.XPATH, "//div[@class='DC_117_pageTabs DC_117_pageActive']")
                active_page_number = active_page_element.text
                next_page_number = str(int(active_page_number) + 1)  # Convert the page number to int, add 1, and convert back to str
                
                # Locate the next page button
                next_page_button_xpath = f"//div[@class='DC_117_paginationTable']//div[text()='{next_page_number}']"
                next_page_button = wait_A.until(EC.presence_of_element_located((By.XPATH, next_page_button_xpath)))
                
                # Scroll into view and click using JavaScript if necessary
                driver_A.execute_script("arguments[0].scrollIntoView(true);", next_page_button)  # Ensure element is in view before clicking
                time.sleep(1)  # Short delay to ensure scroll action is completed
                
                try:
                    next_page_button.click()
                except ElementNotInteractableException:
                    driver_A.execute_script("arguments[0].click();", next_page_button)  # Gracefully handle non-interactive elements
                
                print(f"Navigating to page {next_page_number}")
                time.sleep(10)  # Wait for the page to load new content
            except (NoSuchElementException, TimeoutException):
                print("No more pages to paginate or pagination element not found")
                break
        except Exception as e:
            print(f"Error occurred: {str(e)}")
            break
    
    return LINKS_KERALA, ROUTE_KERALA

# Retrieve links and routes
LINKS_KERALA, ROUTE_KERALA = kerala_link_route("//a[@class='route']")  # Corrected the function name

# Close the WebDriver after scraping
driver_A.quit()  # Corrected 'driver.quit()' to 'driver_A.quit()' as the driver is named 'driver_A'
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException, TimeoutException, ElementNotInteractableException
import time
import pandas as pd

# Initialize WebDriver
driver_A = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
driver_A.maximize_window()

# Open the desired webpage
driver_A.get("https://www.redbus.in/online-booking/jksrtc")
time.sleep(3)  # Allow time for the page to load

# Function to retrieve bus links and routes
def jk_link_route(path):
    LINKS_JK = []
    ROUTE_JK = []
    
    # WebDriverWait instance
    wait_A = WebDriverWait(driver_A, 10)
    # Retrieve route links from multiple pages
    while True:
        try:
            # Extract route links and texts
            paths = driver_A.find_elements(By.XPATH, path)
            for links in paths:
                d = links.get_attribute("href")
                if d:
                    LINKS_JK.append(d)
            
            for route in paths:
                ROUTE_JK.append(route.text)
            
            try:
                # Get the active page number
                active_page_element = driver_A.find_element(By.XPATH, "//div[@class='DC_117_pageTabs DC_117_pageActive']")
                active_page_number = active_page_element.text
                next_page_number = str(int(active_page_number) + 1)  # Converting the page number str to int +1 then converting to str
                
                # Locate the next page button
                next_page_button_xpath = f"//div[@class='DC_117_paginationTable']//div[text()='{next_page_number}']"
                next_page_button = wait_A.until(EC.presence_of_element_located((By.XPATH, next_page_button_xpath)))
                
                # Scroll into view and click using JavaScript if necessary
                driver_A.execute_script("arguments[0].scrollIntoView(true);", next_page_button)  # To ensure element is in view before clicking
                time.sleep(1)  # Short delay to ensure scroll action is completed
                
                try:
                    next_page_button.click()
                except ElementNotInteractableException:
                    driver_A.execute_script("arguments[0].click();", next_page_button)  # If scrollIntoView doesn't work, use JavaScript click
                    
                print(f"Navigating to page {next_page_number}")
                time.sleep(10)  # Wait for the page to load new content
            
            except (NoSuchElementException, TimeoutException):
                print("No more pages to paginate or pagination element not found")
                break
            
        except Exception as e:
            print(f"Error occurred: {str(e)}")
            break
    
    return LINKS_JK, ROUTE_JK

# Retrieve links and routes
LINKS_JK, ROUTE_JK = jk_link_route("//a[@class='route']")

# Close the WebDriver after scraping
driver_A.quit() 

# Create a DataFrame to store the scraped data
df_jk = pd.DataFrame({"Route_name": ROUTE_JK, "Route_link": LINKS_JK})

# Display the DataFrame with bus routes and links
df_jk
```

```python
#save the file in csv format
path = r"/Users/priyankamalavade/Desktop/redbus_scraping/scraped bus details/df.jk.csv"
df_jk.to_csv(path, index=False)
```
